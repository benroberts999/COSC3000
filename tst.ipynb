# Plot weather data

The data is available from bom:
* http://www.bom.gov.au/climate/maps/averages/temperature/

For example, files for maximum/minimum 
* Annual maximum/minimum temperatures averaged over last 30 years:
  * http://www.bom.gov.au/web01/ncc/www/climatology/temperature/mxt/mxtan.zip
  * http://www.bom.gov.au/web01/ncc/www/climatology/temperature/mnt/mntan.zip

* Same, but for January (and so on)
  * http://www.bom.gov.au/web01/ncc/www/climatology/temperature/mxt/mxtjan.zip
  * http://www.bom.gov.au/web01/ncc/www/climatology/temperature/mnt/mntjan.zip

Contains temperature values, for a range of longtitudes and lattitudes.

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
```

### Extract data from file
Unzip manually (or using pythons unzip function).
The first 6 lines are header info

```python
max_data = np.genfromtxt("maxann.txt", skip_header=6)
min_data = np.genfromtxt("minann.txt", skip_header=6)
```

### Plot heatmap
* Seaborn is a nice library for this

```python
ax = sns.heatmap(max_data)
plt.show()
```

This looks bad, because BOM uses value '-9999.' to indicate 'no data'.

We change these to 'nan', since the plot will simply ignore these.

(Note: we are doing float comparison here, which is sometimes not ideal. We could use an "approx =", but it works, so it works.)

```python
# Find all values that = -9999., replace them with nan (not a number)
max_data[max_data == -9999.] = np.nan
min_data[min_data == -9999.] = np.nan
```

```python
ax = sns.heatmap(max_data, cmap = "icefire")
ax.axis('off')
plt.show()
```

```python
ratio = max_data-min_data
ax = sns.heatmap(ratio, cmap = "icefire")
ax.axis('off')
plt.show()
```

```python
months = [datetime(2000, i + 1, 1).strftime("%b") for i in range(12)]
print(months)
```

```python
xllcorner=112
yllcorner=-44
cellsize=0.025

x = [xllcorner + i*cellsize for i in range(len(max_data[0]))]
y = [yllcorner + i*cellsize for i in range(len(max_data))]
y.reverse()
print(x)
print(y)
```

```python
import pandas as pd

df = pd.DataFrame(max_data, columns=x, index=y)

ax = sns.heatmap(df, cmap = "icefire")
ax.annotate("B",[150,-27])
plt.show()
```

```python
xy = [[xx,yy] for xx in x for yy in y]

print(xy)
```
